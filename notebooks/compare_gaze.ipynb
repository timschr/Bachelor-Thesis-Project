```python
import pysaliency
import scipy.io as sio
import numpy as np
import cv2
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt 
```

```python
video_name = 'bridge_2' #'breite_strasse'
molinSaliency_path = '/Users/timschroder/Documents/Uni/Bachelorarbeit/project_code/proto-objects/molinSaliency/' + video_name + '.mat'
mySaliency_path = '/Users/timschroder/Documents/Uni/Bachelorarbeit/project_code/proto-objects/mySaliency/' + video_name + '.mat'
waltherSaliency_path = '/Users/timschroder/Documents/Uni/Bachelorarbeit/project_code/proto-objects/waltherSaliency/' + video_name + '.mat'
```

```python
molinSaliency_raw = sio.loadmat(molinSaliency_path)
mySaliency_raw = sio.loadmat(mySaliency_path)
waltherSaliency_raw = sio.loadmat(waltherSaliency_path)

```

```python
molinSaliency = np.asarray(molinSaliency_raw['molin_sal_maps'])     #.shape = (450, 800, 596)
mySaliency = np.asarray(mySaliency_raw['video_saliency'])           #.shape = (28, 50, 596)
waltherSaliency = np.asarray(waltherSaliency_raw['video_saliency']) #.shape = (28, 50, 596)
```

```python
f = 200
my_resized = cv2.resize(mySaliency[:,:,f], (800,450), interpolation = cv2.INTER_AREA)
my_blur = cv2.blur(my_resized,(20,20))
walther_resized = cv2.resize(waltherSaliency[:,:,f], (800,450), interpolation = cv2.INTER_AREA)
walther_blur = cv2.blur(walther_resized,(20,20))
```
